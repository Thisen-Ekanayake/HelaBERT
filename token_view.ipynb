```python
import sentencepiece as spm

MODEL_PATH = "tokenizer/unigram_32000_0.9995.model"

sp = spm.SentencePieceProcessor()
sp.load(MODEL_PATH)

vocab_size = sp.get_piece_size()
print(f"Vocab size: {vocab_size}\n")

print("First 100 token IDs and tokens:\n")

for token_id in range(min(100, vocab_size)):
    token = sp.id_to_piece(token_id)
    score = sp.get_score(token_id)
    print(f"{token_id:4d} | {token} | score={score:.6f}")
```
